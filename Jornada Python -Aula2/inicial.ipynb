### Case - Cancelamento de Clientes

Você foi contratado por uma empresa com mais de 800 mil clientes para um projeto de Dados. Recentemente a empresa percebeu que da sua base total de clientes, a maioria são clientes inativos, ou seja, que já cancelaram o serviço.

Precisando melhorar seus resultados ela quer conseguir entender os principais motivos desses cancelamentos e quais as ações mais eficientes para reduzir esse número.

Base de dados e arquivos: https://drive.google.com/drive/folders/1uDesZePdkhiraJmiyeZ-w5tfc8XsNYFZ?usp=drive_link

```python
# passo a passo do projeto
# pip install pandas openpyxl ipykernel nbformat plotly

# passo1: abrir a base de dados(importar  a base de dados)
import pandas as pd #as pd = escrever menos, pd é o apelido do pandas.
tabela = pd.read_csv("cancelamentos.csv") # o cara da esquerda recebe o valor da direita.
display(tabela)

# passo2: visualizar a base de dados
    #entender as info disponiveis
    #possiveis problemas/erros na base de dados

    #kaggle.com é um site (treinamento, que disbonibiliza dados para codigos)
    #informções que não te ajudam - atrapalham.

tabela = tabela.drop(columns="CustomerID")
display(tabela)
```

```python
# passo 3: corrigir os problemas da base de dados(tratamento de dados)
display(tabela.info()) #info() = informações da tabela

    # problemas que vamos ter que resolver:
        #valores em formatos errados -> OK
        #valores faltando(vazias)
        
tabela = tabela.dropna() #dropna() = apagar valores vazios not a number(NaN)
```

```python
# passo 4: analise inicial (entender quantos clientes cancelaram)

# quantidade - contar quantos caras são 0 e quantos são 1 na coluna cancelou
tabela["cancelou"].value_counts() #value_counts() = contar quantos valores existem na coluna cancelou

# porcentagem
display(tabela["cancelou"].value_counts(normalize=True)) #normalize=True = porcentagem  

# display(tabela["cancelou"].value_counts(normalize=True).map("{:.2%}".format)) 
```

```python
#passo 5: analise detalhada (causa do cancelamento dos clientes, como cada coluna impacta no cancelamento)
import plotly.express as px #importar a biblioteca plotly.express como px

print(tabela.columns)


for coluna in tabela.columns:
    # cria o grafico
    grafico = px.histogram(tabela, x=coluna, color="cancelou", text_auto=True)
    # exibe o grafico
    grafico.show()

#curso streamlit hashtag (video)
```

```python
#1.todo mundo do "contrato mensal", cancelou o serviço
    #vamos dar desconto no contrato anual, trimestral

#2.ligações_callcenter acima de 4, todo mundo cancelou
    #tem algum problema que não estamos conseguindo resolver
    #se o cliente ligar 3 pro callcenter, alerta vermelho

#3. atraso no pagamento acima de 20 dias, o cliente cancela
    #se o cliente atrasar 15 dias no pagamento, alerta vermelho

#cancelamento = 56%

# filtrar a base de dados
# duracao_contrato -> diferente de Monthly
condicao = tabela["duracao_contrato"] != "Monthly"
tabela = tabela[condicao]

# ligacoes_contrato -> menores ou iguais a 4
condicao = tabela["ligacoes_callcenter"] <= 4
tabela = tabela[condicao]

# dias_atraso -> menores ou iguais a 20
condicao = tabela["dias_atraso"] <= 20
tabela = tabela[condicao]

# porcentagem
display(tabela["cancelou"].value_counts(normalize=True)) #normalize=True = porcentagem  
```
